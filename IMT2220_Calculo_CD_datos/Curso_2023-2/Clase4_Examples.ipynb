# Funciones en varias variables

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import Normalize
```

```python
def f1(x, y):
    return x + y*(x + y)

def f2(x):     
    return x[0] + x[1]*(x[0] + x[1])

f3 = lambda x, y: x + y*(x + y)
```

```python
t = np.linspace(-1, 1)

xx, yy = np.meshgrid(t, t)

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection':'3d'})

ax.plot_surface(xx, yy, f1(xx, yy), antialiased=True)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={"projection": "3d"})

surf = ax.plot_surface(xx, yy, f1(xx, yy), cmap='jet')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## Paraboloide elíptico

```python
t = np.linspace(-1, 1)

xx, yy = np.meshgrid(t, t)

paraboloide = lambda x, y: x**2 + y**2

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection':'3d'})

ax.plot_surface(xx, yy, paraboloide(xx, yy), antialiased=True)

plt.show()
```

## Radial Function

```python
t = np.linspace(-1, 1)

xx, yy = np.meshgrid(t, t)

radial = lambda x, y: np.sqrt(x**2 + y**2)

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection':'3d'})

ax.plot_surface(xx, yy, radial(xx, yy), antialiased=True)

plt.show()
```

## Paraboloide hiperbólico

```python
t = np.linspace(-1, 1)

xx, yy = np.meshgrid(t, t)

hyperbolic = lambda x, y: x**2 - y**2

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection':'3d'})

ax.plot_surface(xx, yy, hyperbolic(xx, yy), antialiased=True)

ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

## Otras funciones extrañas

```python
t = np.linspace(-1, 1)

xx, yy = np.meshgrid(t, t)

f = lambda x, y: np.sin(np.pi*x)*y**2 + x*np.cos(np.pi*y)

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={'projection':'3d'})

ax.plot_surface(xx, yy, f(xx, yy), antialiased=True)

plt.show()
```

```python
%matplotlib

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')

# Make data
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))


def T(x, y, z):
    if z > 0:
        return np.random.normal(loc=35, scale=2) - 4*z
    return  np.random.normal(loc=30, scale=1) + z - min([z**2, 30]) 

vecT = np.vectorize(T)

C = vecT(x, y, z)
cmap = cm.coolwarm
norm = Normalize(vmin=np.amin(C), vmax=np.amax(C))


# Plot the surface
surf = ax.plot_surface(x, y, z, facecolors=cmap(norm(C)), cmap='coolwarm', alpha=0.9, linewidth=0)

m = cm.ScalarMappable(cmap=cm.coolwarm)
m.set_array(vecT(x, y, z))
cbar = plt.colorbar(m, shrink=0.6)

ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

plt.title(r"Temperatura $(^\circ C)$ alrededor de una esfera de radio $r=10$")

plt.show()
```

## Contour plots

```python
fig = plt.figure(figsize=(8, 8))

CS = plt.contourf(xx, yy, f1(xx, yy), levels=10, cmap='jet')
cbar = fig.colorbar(CS, label='value')

CS2 = plt.contour(CS, levels=CS.levels[::1], colors='k', alpha=0.5)
cbar.add_lines(CS2)

plt.show()
```

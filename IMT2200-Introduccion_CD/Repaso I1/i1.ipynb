### Normalizacion de un JSON a un DataFrame:


``` python

df = pd.json_normalize(json, sep=".", record_path="column", meta=["col1", "col2"])

```

`record_path` y `meta` son dos parámetros opcionales de la función `json_normalize()` de Pandas que se utilizan para normalizar JSONs con estructuras anidadas.

- `record_path`: indica la ruta a la tabla anidada que se debe normalizar. Si el JSON tiene una estructura de tabla anidada, `record_path` se utiliza para especificar la ruta a la tabla anidada. Por ejemplo, si el JSON tiene una estructura como esta:



````python
{
    "col1": "valor1",
    "col2": "valor2",
    "column": [
        {"a": 1, "b": 2},
        {"a": 3, "b": 4},
        {"a": 5, "b": 6}
    ]
}
````



y queremos normalizar la tabla anidada en la clave "column", podemos usar `record_path="column"`.

- `meta`: indica las claves adicionales del JSON que se deben agregar como columnas en el data frame resultante. Por ejemplo, si el JSON tiene las claves "col1" y "col2" que queremos agregar como columnas en el data frame, podemos usar `meta=["col1", "col2"]`.

En resumen, `record_path` se utiliza para especificar la ruta a la tabla anidada que se debe normalizar, mientras que `meta` se utiliza para agregar claves adicionales del JSON como columnas en el data frame resultante.

#### Ejemplo de normalizacion

````python
json_list = [
    { 
        'class': 'Year 1', 
        'student count': 35, 
        'room': 'A2',
        'info': {
            'teachers': { 
                'math': 'Emmy Noether', 
                'physics': 'Richard Feynman' 
            }
        },
        'students': [
            { 
                'name': 'Mary', 
                'sex': 'F', 
                'grades': { 'math': 75, 'physics': 98 } 
            },
            { 
                'name': 'James', 
                'sex': 'M', 
                'grades': { 'math': 80, 'physics': 78 } 
            },
        ]
    },
    { 
        'class': 'Year 2', 
        'student count': 28, 
        'room': 'A4',
        'info': {
            'teachers': { 
                'math': 'Alan Turing', 
                'physics': 'Vera Rubin' 
            }
        },
        'students': [
            { 'name': 'Tony', 'sex': 'M' },
            { 'name': 'Jacqueline', 'sex': 'F' },
        ]
    },
]
all_students_grades= pd.json_normalize(json_list, sep="_", record_path="students", meta=["class", "room", ["info"], ["info", "teachers", "physics"]])
````

Notemos que con el comando ```record_path``` descomprimimos el ```key``` que contiene a los estudiantes, y a su vez con ```meta``` le especificamos a pandas que agregue las columnas con la informacion, aunque notemos que dentro de meta colocamos una lista ```["info", "teachers", "math"]``` este orden tambien descomprime el diccionario.

````python
all_students_grades
````

### Extraccion de datos de una pagina web:

Para extraer datos de una página web con BeautifulSoup (bs4), primero debes obtener el HTML de la página web. Esto se puede hacer utilizando la biblioteca `requests` de Python para enviar una solicitud HTTP a la página web y obtener su contenido.

Una vez que tengas el HTML, puedes crear un objeto `BeautifulSoup` a partir del HTML para analizarlo y extraer los datos que necesitas. Puedes usar los métodos `find()` y `find_all()` de BeautifulSoup para buscar elementos HTML específicos en el HTML y extraer su contenido.

Por ejemplo, si quieres extraer el texto de todos los elementos `<p>` en una página web, puedes hacer lo siguiente:



````python
import requests
from bs4 import BeautifulSoup

# Obtener el HTML de la página web
url = "https://www.example.com"
response = requests.get(url)
html = response.content

# Crear un objeto BeautifulSoup a partir del HTML
soup = BeautifulSoup(html, "html.parser")

# Encontrar todos los elementos <p> y extraer su texto
paragraphs = soup.find_all("p") # Ademas podemos buscar "table", "tr", "tb"
for p in paragraphs:
    print(p.text)
````

#### Ejemplo de extraccion de datos de una pagina web:


````python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_coffee_production"
page = requests.get(url).text
soup = bs(page, "html.parser").find_all("table")[0] # Ordenar la informacion
````

````python
df = pd.DataFrame(columns=['country', 'bags_60kg', 'tons_metric','pounds'])
for row in soup.find_all('tr')[1::]:
    cols = [col.text.strip() for col in row.find_all("td")]
    country, bags, touns, pounds = cols[1], cols[2], cols[3], cols[4]
    new_row = pd.DataFrame({'country': country, 'bags_60kg': bags, 'tons_metric': tons,'pounds':pounds}, index=['country'])
    df = pd.concat([df, new_row], ignore_index=True)
df
````

### Grafica de dataframes

Para graficar un DataFrame en un gráfico scatter con Matplotlib y Seaborn, puedes utilizar las funciones `scatter()` de Matplotlib.

Aquí hay un ejemplo de cómo graficar un DataFrame en un gráfico scatter con Matplotlib:



````python
# Crear un DataFrame de ejemplo
data = {'x': [1, 2, 3, 4, 5],
        'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

# Graficar el DataFrame en un gráfico scatter con Matplotlib
plt.scatter(df['x'], df['y'])
plt.show()
````

Ademas para graficar un DataFrame en un gráfico de líneas con Matplotlib, puedes utilizar la función `plot()` de Matplotlib.


````python
# Crear un DataFrame de ejemplo
data = {'x': [1, 2, 3, 4, 5],
        'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)

# Graficar el DataFrame en un gráfico de líneas con Matplotlib
plt.plot(df['x'], df['y'])
plt.show()
````

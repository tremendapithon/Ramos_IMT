## 2. Relaciones

Queremos ver si la relacion entre las variables medidas ha ido cambiando en el tiempo. En la Interrogación vimos que el mejor modelo era el multilineal, así que aprovecharemos eso para entender cómo cambia la relación entre las variables en el tiempo. ### 2.1 Modelos anuales
Para cada año, haga un modelo de regresión multilineal que considere la `Salinidad` como variable dependiente y las mediciones de `Temperatura`, `Profundidad` y `Densidad` como las variables independientes. No necesita separar los datos entre prueba y validación para este ejercicio.

En cada modelo, calcule el r², el RMSE y almacene los valores de $\beta_0$, $\beta_1$, $\beta_2$ y $\beta_3$ (los coeficientes de la regresión).

### 2.2 Visualización
Haga los siguientes gráficos:
- Un único gráfico que muestre, para cada año, el valor de los coeficientes de la regresión multilineal ($\beta_0$, $\beta_1$, $\beta_2$ y $\beta_3$). ¿Se mantienen estables a lo largo de los años?
- Un gráfico que muestre el valor de r² para cada año.
- Un gráfico que muestre el valor del RMSE para cada año. ¿Cambia el nivel de error a lo largo de los años?

```python
# librerias necesarias
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# se crea el modelo
r2 = list()
coef = list()
rmse = list()

for year in oceanData.Year.unique():
    model = LinearRegression()
    x = np.array(oceanData[oceanData.Year == year][['T_degC', "Depthm", "STheta"]])
    y = np.array(oceanData[oceanData.Year == year]['Salnty'])
    train = model.fit(x, y)
    pred = model.predict(x)

    # Almacenamos po
    r2.append([year, model.score(x, y)])
    coef.append([year, model.coef_])
    rmse.append([year, mean_squared_error(y, pred)])
    
```

```python
# Grafico coef
coef = pd.DataFrame(coef, columns=["Year", "Coef"])
co = pd.DataFrame(coef.Coef.values.tolist(), columns=["T_degC", "Depthm", "STheta"])
coef = pd.concat([coef.Year, co], axis=1)
coef
```

```python
# Grafico de los coeficientes
sns.lineplot(data=coef, x="Year", y="T_degC", label="T_degC")
sns.lineplot(data=coef, x="Year", y="Depthm", label="Depthm")
sns.lineplot(data=coef, x="Year", y="STheta", label="STheta")
```

El unico valor que se mantiene constante al paso del tiemoi es la profundidad.

```python
# Grafico de R2
r2 = pd.DataFrame(r2, columns=["Year", "R2"])
plt.figure(figsize=(10, 10), dpi=200)
sns.lineplot(data=r2, x="Year", y="R2")
```

```python
# Grafico de RMSE
rmse = pd.DataFrame(rmse, columns=["Year", "RMSE"])
plt.figure(figsize=(10, 10), dpi=200)
sns.lineplot(data=rmse, x="Year", y="RMSE")
```

Donde existen anomalias en la MSE es en los años 2010 y 2020, donde se observa un aumento en el error, incluso en los graficos antes mostrados se observan anomalias en los años 2010 y 2020.
